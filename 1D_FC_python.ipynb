```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 401
D = np.linspace(0, 1, N).reshape(-1, 1)
phi_D = np.sin(np.pi*D)
```

```python
C = np.fft.fft(phi_D, axis=0) / len(phi_D)
C.shape
```

```python
D_err = np.linspace(0, 1, N*20+1).reshape(-1, 1)
D_err.shape
```

```python
if N%2 == 0:
    freq = np.concatenate((np.arange(N//2+1), np.arange(-N//2+1, 0))).reshape(-1, 1)
else:
    freq = np.concatenate((np.arange((N-1)//2+1), np.arange(-(N-1)//2,0))).reshape(-1, 1)
freq.shape
```

```python
phi_FT = np.real(np.exp(2j*np.pi*D_err@freq.T) @ C)
phi_FT.shape
```

```python
plt.figure()
plt.plot(D_err, phi_FT)
plt.plot(D_err, np.sin(np.pi*D_err))
```

```python
np.max(phi_FT - np.sin(np.pi*D_err))
```
